```python
img_height, img_width = 256,256
```

```python
train_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255, validation_split=0.2)
SEED = 2002
root_dir = "data"

train_set = train_gen.flow_from_directory(
    "images",
    target_size=(img_height, img_width),
    class_mode='categorical',
    batch_size=256,
    shuffle=True,
    seed=SEED,
    subset='training'
)

valid_set = train_gen.flow_from_directory(
    "images",
    target_size=(img_height, img_width),
    class_mode='categorical',
    batch_size=256,
    shuffle=False,
    subset='validation'
)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation='relu', padding = "same", input_shape=(img_height, img_width, 3)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(43, activation='softmax')
])
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# num_epochs = 5
# history = model.fit(train_set, epochs=num_epochs, validation_data=valid_set)
```
